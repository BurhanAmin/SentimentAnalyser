```python
import torch
from transformers  import DistilBertTokenizer
data["text"] = data["text"].astype(str)
tokens = tokenizer(
    data["text"].tolist(),
    padding = True,
    truncation = True,
    max_lenght = 512,
    return_tensors = "pt"
)
```

```python
sentiment_maping = {
    'negative':0,
    'neutral':1,
    'postive':2

}
data["sentiment"]=data["sentiment"].map(sentiment_maping)

labels = torch.tensor(data["sentiment"].values)
```

```python
torch.save((tokens, labels) , "tokenized_dataset.pt")
```
